## Other/Further Analysis

## Men vs. Women in Each Category Major
We use a grouped bar plot, making a new data frame first with counts

```python
recent_grads['Major_category'].unique()
```

```python
categories = ['Engineering', 'Business', 'Physical Sciences',
       'Law & Public Policy', 'Computers & Mathematics',
       'Agriculture & Natural Resources',
       'Industrial Arts & Consumer Services', 'Arts', 'Health',
       'Social Science', 'Biology & Life Science', 'Education',
       'Humanities & Liberal Arts', 'Psychology & Social Work',
       'Communications & Journalism', 'Interdisciplinary']
men_count = []
women_count = []
for cat in categories:
    unique_category = recent_grads[recent_grads['Major_category'] == cat]
    sum_men = unique_category['Men'].sum()
    sum_women = unique_category['Women'].sum()
    men_count.append(sum_men)
    women_count.append(sum_women)

recent_grads_genders_category = pd.DataFrame({'Men': men_count, 'Women': women_count},
                                             index = categories)

recent_grads_genders_category.plot.bar(legend=True, figsize = (10,10))
```

We see that engineering has a lot more men graduating than women as expected, as well as mathematics. Women outnumber men in places like education, the arts, and health. Business, the highest category, has an even distribution of both, which is good.
